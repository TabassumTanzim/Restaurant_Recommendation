```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import numpy as np

path = r"/Users/tabassumtanzim/Downloads/chromedriver-mac-arm64/chromedriver"

# Create Chrome object
options = webdriver.ChromeOptions()
options.add_argument(f"webdriver.chrome.driver={path}")

driver = webdriver.Chrome(options=options)
driver.get("https://www.foodpanda.com.bd/restaurants/new/?lat=23.71501&lng=90.41864&vertical=restaurants&expedition=delivery&query=wari+restuarant")



title=[]
cuisine=[]
rating=[]
location='Wari'


for i in range (1,33):
    if (driver.find_element(By.XPATH, '//*[@id="restaurant-listing-root"]/div/div/div[2]/div[2]/div/div/div/div[4]/section/ul/li[' + str(i) + ']/div/a/figure/figcaption/span/span[1]').text != 'Top restaurant'):
        try:
            title.append(driver.find_element(By.XPATH, '//*[@id="restaurant-listing-root"]/div/div/div[2]/div[2]/div/div/div/div[4]/section/ul/li[' + str(i) + ']/div/a/figure/figcaption/span/span[1]').text)
        except:
            title.append(driver.find_element(By.XPATH, '//*[@id="restaurant-listing-root"]/div/div/div[2]/div[2]/div/div/div/div[4]/section/ul/li[' + str(i) + ']/div/a/figure/figcaption/span/span[2]').text)

                                    
        cuisine.append(driver.find_element(By.XPATH, '//*[@id="restaurant-listing-root"]/div/div/div[2]/div[2]/div/div/div/div[4]/section/ul/li[' + str(i) + ']/div/a/figure/figcaption/ul[1]/li[2]/span').text)
      
        try:
            rating.append(driver.find_element(By.XPATH,'//*[@id="restaurant-listing-root"]/div/div/div[2]/div[2]/div/div/div/div[4]/section/ul/li[' + str(i) + ']/div/a/figure/figcaption/span/span[2]/div/div/span[1]').text)
        except:
            rating.append(driver.find_element(By.XPATH,'//*[@id="restaurant-listing-root"]/div/div/div[2]/div[2]/div/div/div/div[4]/section/ul/li[' + str(i) + ']/div/a/figure/figcaption/span/span[3]/div/div/span[1]').text)
              

                          
print(title)        
print(cuisine)
print(rating)



```

```python
data = {'Location':location,'Restuarants': title, 'Cuisine': cuisine, 'Rating': rating}
df=pd.DataFrame(data)
df
```
